```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("raw-data.xlsx", header=1)
df
```

# Drop empty Rows

```python
df = df.dropna(subset=["Student Name"])
df
```

# Convert Columns in correct Datatypes
## Team
* Set Professor's Example to Team 0
* Convert Column to Numeric Datatype

```python
df.loc[0, "Team "] = 0
df = df.astype({"Team ": "int8"})
df
```

## Studnet Number
* Set Professor's Student Number to 0
* Remove "Student" (e.g. Student 3 -> 3)
* Convert Column to Numeric Datatype

```python
df.loc[0, "Student Number"] = "Student 0"
df.loc[:, "Student Number"] = df["Student Number"].str.replace("Student ", "")
df = df.astype({"Student Number": "int8"})
df
```

## Weight (lbs)
* Remove "lbs"
* Convert Column to Numeric Datatype

```python
df.loc[:, "Weight (lbs)"] = (
    df["Weight (lbs)"].astype(str).str.replace("lbs", "", regex=False)
)
df = df.astype({"Weight (lbs)": "float32"})
df
```

## % Weight distribution in front
* Remove --, -, None listed, and Not listed from Column
* Convert fractions (e.g. 60/40 -> 60)
* Convert Column to Numeric Datatype

```python
def clean_value(val):
    if str(val).strip() in ["--", "-", "None listed", "Not Listed"]:
        return np.nan

    if isinstance(val, str) and "/" in val:
        numerator = val.split("/")[0]
        return numerator

    return val


df.loc[:, "% Weight distribution in front"] = df[
    "% Weight distribution in front"
].apply(clean_value)
df = df.astype({"% Weight distribution in front": "float32"})
df
```

## Length / Width / Height
* Remove "in"
* Convert Columns to Numeric Datatype

```python
df.loc[:, ["Length", "Width", "Height"]] = (
    df[["Length", "Width", "Height"]].astype(str).map(lambda x: x.replace("in", ""))
)
df = df.astype({"Length": "float32", "Width": "float32", "Height": "float32"})
df
```

## Engine Displacement (L)
* Remove "L"
* Convert Column to Numeric Datatype

```python
df.loc[:, "Engine Displacement (L)"] = (
    df["Engine Displacement (L)"].astype(str).str.replace("L", "", regex=False)
)
df = df.astype({"Engine Displacement (L)": "float32"})
df
```

# Drop Rows
## Transmission (Speeds) Listed as S6 in the "TRANS" column
* Remove transmission type labels (e.g., Manual, Automatic) and only keep gear count number (e.g., Auto(M6) → 6)
* Keep AV Engines (since I don't know what it is it needs further investigation)

```python
def clean_data(val):
    if pd.isna(val):
        return np.nan

    if "AV" in str(val):
        return "AV"

    if isinstance(val, (int, float)):
        return val

    digits = [digit for digit in str(val) if str.isdigit(digit)]
    return int(digits[0]) if digits else np.nan


df.loc[:, 'Transmission (Speeds) Listed as S6 in the "TRANS" column'] = df[
    'Transmission (Speeds) Listed as S6 in the "TRANS" column'
].apply(clean_data)
df
```

# Year / Weight Investigation
### Little bit weird that Weight Range and the Weight (lbs) don't align (same for Year Range and Model Year)

```python
matching_years, not_matching_years = [], []
matching_weights, not_matching_weights = [], []

for idx, row in df.iterrows():
    year_from, year_to = (
        int(row["Year Range"].split("-")[0]),
        int(row["Year Range"].split("-")[1]),
    )
    weight_from, weight_to = (
        int(row["Weight Range"].split("-")[0]),
        int(row["Weight Range"].split("-")[1]),
    )

    if year_from <= row["Model Year"] <= year_to:
        matching_years.append(idx)
    else:
        not_matching_years.append(idx)

    if weight_from <= row["Weight (lbs)"] <= weight_to:
        matching_weights.append(idx)
    else:
        not_matching_weights.append(idx)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

axs[0].bar(
    ["Matching Years", "Not Matching Years"],
    [len(matching_years), len(not_matching_years)],
    color=["green", "red"],
)
axs[0].set_title("Matching Years vs Not Matching Years")
axs[0].set_ylabel("Count")
axs[0].set_ylim(0, len(df))

axs[1].bar(
    ["Matching Weights", "Not Matching Weights"],
    [len(matching_weights), len(not_matching_weights)],
    color=["green", "red"],
)
axs[1].set_title("Matching Weights vs Not Matching Weights")
axs[1].set_ylabel("Count")
axs[1].set_ylim(0, len(df))

plt.tight_layout()
plt.show()
```

```python
df_team_sort = df.sort_values(by=["Team", "Student Number"])
df_student_sort = df.sort_values(by=["Student Number", "Team"])
```

```python
cmap = plt.get_cmap("tab20")
n_colors = cmap.N

ctr = 0
year_idx_ranges, year_ranges, year_colors = [], [], []
for i, team in enumerate(df_team_sort["Team"].unique()):
    df_temp = df_team_sort[df_team_sort["Team"] == team]
    year_idx_ranges.append((ctr, ctr + len(df_temp) - 1))
    year_ranges.append(
        (
            int(df_temp.iloc[0]["Year Range"].split("-")[0]),
            int(df_temp.iloc[0]["Year Range"].split("-")[1]),
        )
    )
    year_colors.extend([cmap(i % n_colors)] * len(df_temp))
    ctr += len(df_temp)

fix, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

axs[0].scatter(np.arange(len(df)), df_team_sort["Model Year"], c=year_colors)
for i, ((idx_from, idx_to), (year_from, year_to)) in enumerate(
    zip(year_idx_ranges, year_ranges)
):
    if i == 0:
        axs[0].hlines(
            year_from, idx_from, idx_to, colors="black", linewidth=2, label="Year Range"
        )
    else:
        axs[0].hlines(year_from, idx_from, idx_to, colors="black", linewidth=2)

    axs[0].hlines(year_to, idx_from, idx_to, colors="black", linewidth=2)

axs[0].set_title("Model Year by Team")
axs[0].set_xlabel("Index (Sorted by Team)")
axs[0].set_ylabel("Model Year")
axs[0].legend()

axs[1].hist(df["Model Year"], bins=10, edgecolor="black")
axs[1].set_title("Distribution of Model Year")
axs[1].set_xlabel("Model Year")
axs[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
cmap = plt.get_cmap("tab20")
n_colors = cmap.N

ctr = 0
weight_idx_ranges, weight_ranges, weight_colors = [], [], []
for i, student in enumerate(df_student_sort["Student Number"].unique()):
    df_temp = df_student_sort[df_student_sort["Student Number"] == student]
    weight_idx_ranges.append((ctr, ctr + len(df_temp) - 1))
    weight_ranges.append(
        (
            int(df_temp.iloc[0]["Weight Range"].split("-")[0]),
            int(df_temp.iloc[0]["Weight Range"].split("-")[1]),
        )
    )
    weight_colors.extend([cmap(i % n_colors)] * len(df_temp))
    ctr += len(df_temp)

fix, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

axs[0].scatter(np.arange(len(df)), df_student_sort["Weight (lbs)"], c=weight_colors)
for i, ((idx_from, idx_to), (weight_from, weight_to)) in enumerate(
    zip(weight_idx_ranges, weight_ranges)
):
    if i == 0:
        axs[0].hlines(
            weight_from,
            idx_from,
            idx_to,
            colors="black",
            linewidth=2,
            label="Weight Range",
        )
    else:
        axs[0].hlines(weight_from, idx_from, idx_to, colors="black", linewidth=2)

    axs[0].hlines(weight_to, idx_from, idx_to, colors="black", linewidth=2)

axs[0].set_title("Model Weight by Student Number")
axs[0].set_xlabel("Index (Sorted by Student)")
axs[0].set_ylabel("Weight (lbs)")
axs[0].legend()

axs[1].hist(df["Weight (lbs)"], bins=10, edgecolor="black")
axs[1].set_title("Distribution of Weight (lbs)")
axs[1].set_xlabel("Weight (lbs)")
axs[1].set_ylabel("Frequency")

plt.show()
```
