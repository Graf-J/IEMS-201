```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned-data.csv", index_col=0)
df
```

```python
pd.plotting.scatter_matrix(
    df[
        [
            "Model Year",
            "Weight (lbs)",
            "Weight Distribution (%)",
            "Length (in)",
            "Width (in)",
            "Height (in)",
            "Cylinders",
            "Engine Displacement (L)",
            "Transmission Type",
            "Transmission Speeds",
            "MPG",
        ]
    ],
    alpha=0.8,
    figsize=(20, 20),
    diagonal="kde",
)

plt.tight_layout()
plt.show()
```
