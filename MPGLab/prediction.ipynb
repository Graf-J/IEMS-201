```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("cleaned-data.csv", index_col=0)
df
```

# Impute Missing Values with Mean

```python
cols_with_na = df.columns[df.isnull().any()]
imputer = SimpleImputer(strategy="mean")
df[cols_with_na] = imputer.fit_transform(df[cols_with_na])
df
```

# Multiple Regression (all Parameters)

```python
model = sm.OLS(
    df["MPG"],
    df[
        [
            "Model Year",
            "Weight (lbs)",
            "Weight Distribution (%)",
            "Length (in)",
            "Width (in)",
            "Height (in)",
            "Cylinders",
            "Engine Displacement (L)",
            "Transmission Type",
            "Transmission Speeds",
        ]
    ],
)
res = model.fit()
print(res.summary())
```
